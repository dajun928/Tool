```python
#coding=utf-8
import matplotlib.pyplot as plt
import random
import sys 
reload(sys) # Python2.5 初始化后会删除 sys.setdefaultencoding 这个方法，我们需要重新载入 
sys.setdefaultencoding('utf-8')
import matplotlib
# matplotlib.use('qt4agg')
#指定默认字体
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['font.family']='sans-serif'
#解决负号'-'显示为方块的问题
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 某城市一小时内的温度变化情况
# plt.figure(figsize=(20, 8), dpi=100)
# axes列表
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=100)

# 准备温度数据
x = list(range(0, 60))
y = [random.uniform(15, 18) for i in range(0, 60)]

# 准备北京的温度
y_beijing = [random.uniform(1, 3) for i in range(0, 60)]

# 画折线图
# 添加线条注释，还需要增加显示
axes[0].plot(x, y, label="上海")
axes[1].plot(x, y_beijing, color='r', linestyle='--', label="北京")

x_name = [u"11点{}分".format(i) for i in range(0, 60)]

# 图的名字，标签，美化过程，在show之前使用
# 美化刻度值时候，必须使用数值类型的数字
plt.xticks(x[::5], x_name[::5])
plt.yticks(list(range(0, 45))[::5])

# 美化标题，横坐标描述，纵坐标描述
# 增加折线注释
plt.legend(loc="best")
plt.xlabel(u"时间")
plt.ylabel(u"温度")
plt.title(u"某周11点0分到11点59分的不同地区温度变化情况")

plt.show()
```

```python
# 画条形图显示电影票房对比

plt.figure(figsize=(20, 8), dpi=100)

# 准备数据
movie_name = [u'雷神3：诸神黄昏',u'正义联盟',u'寻梦环游记']
first_day = [10587.6,10062.5,1275.7]
first_weekend=[36224.9,34479.6,11830]
x = range(len(movie_name))


# 显示三部电影第一天的票房
plt.bar(x, first_day, width=0.2, label=u"首日票房")
plt.bar([i+0.2 for i in x], first_weekend, width=0.2, label=u"首周票房")

# 添加图示
plt.legend(loc="best")

# 修改刻度的位置和值
plt.xticks([i+0.1 for i in x], movie_name)

#标题
plt.title(u"每日每周票房对比")

# 显示
plt.show()
```
